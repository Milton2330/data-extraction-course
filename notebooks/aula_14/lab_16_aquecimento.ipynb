## 🛠️ Cenário: O Conversor de Temperatura

Você tem uma base de dados simulando medições de sensores industriais em **Fahrenheit**. Sua missão é criar a classe `ConversorTemperatura` que receba essa base, aplique a fórmula aritmética para criar a conversão em **Celsius** e descarte a coluna antiga.
